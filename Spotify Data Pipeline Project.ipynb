# Authentication:

* Definition: Authentication is the process of verifying the identity of a user or system entity.
* Purpose: It ensures that the user or entity is who they claim to be.
* Example: When you log in to your email account, you typically provide a username and password. 
        The system then verifies that the credentials you provided match those stored in its database. If they match, you are granted access to your email account. Other forms of authentication include biometric methods like fingerprint or facial recognition, smart cards, security tokens, etc.

# Authorization:

* Definition: Authorization is the process of determining what actions or resources a user or system entity is allowed to access or perform.
* Purpose: It controls access to resources or functionalities based on the authenticated identity.
* Example: Suppose you have a file-sharing system. After you log in (authentication), the system checks your user role or permissions to determine what actions you're allowed to take. For instance, a regular user might only be authorized to view files, while an administrator might have permission to create, edit, and delete files. Authorization can be role-based, attribute-based, or rule-based, depending on the system's design.

```python
album_list = []

for row in data['items']:
    
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    album_release_date = row['track']['album']['release_date']
    album_total_tracks = row['track']['album']['total_tracks']
    album_url = row['track']['album']['external_urls']['spotify']
    album_element = {'album_id' : album_id,'album_name':album_name,'release_date':album_release_date,
                             'total_tracks':album_total_tracks,'url':album_url}
    album_list.append(album_element)
```

```python
artist_list = []
for row in data['items']:
    
    for key, value in row.items():
        
        if key == 'track':
            
            for artist in value['artists']:
                
                artist_dict = {'artist_id' : artist['id'],'artist_name' : artist['name'], 'external_url' : artist['href']}
                artist_list.append(artist_dict)
                
```

```python
song_list = []
for row in data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularity = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_element = {'song_id':song_id,'song_name':song_name,'duration_ms':song_duration,'url':song_url,
                    'popularity':song_popularity,'song_added':song_added,'album_id':album_id,
                    'artist_id':artist_id
                   }
    song_list.append(song_element)
```
